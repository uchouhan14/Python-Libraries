## Filtering dataframe on a condition | Masking using boolean series.

```python
mask = data['city']=='Hyderabad'
```

```python
mask
```

```python
data[mask].head(3)
```

```python
data[mask].shape
```

```python
# A Function which inputs the city name and returns more details about it.

def get_city_match_count(city):
    mask=data['city']==city
    return data[mask].shape[0]
```

```python
get_city_match_count('Indore')
```

```python
mask1 =  mask=data['city']=='Hyderabad'
mask2 = data['date'] > '2015-01-01'
data[mask1 & mask2]
# Here we have created 2 masks and then we use bitwise & operator on
# the masks to get matches played in Hyderabad after 2015-01-01.
```
