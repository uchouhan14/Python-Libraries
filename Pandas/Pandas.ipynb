## Filtering dataframe on a condition | Masking using boolean series.

```python
mask = data['city']=='Hyderabad'
```

```python
mask
```

```python
data[mask].head(3)
```

```python
data[mask].shape
```

```python
# A Function that inputs the city name and returns counts about it.

def get_city_match_count(city):
    mask=data['city']==city
    return data[mask].shape[0]
```

```python
get_city_match_count('Indore')
```

```python
mask1 = data['city']=='Hyderabad'
mask2 = data['date'] > '2015-01-01'
data[mask1 & mask2]
# Here we have created 2 masks which are boolean series that's why we use bitwise & operator on
# the masks to get matches played in Hyderabad after 2015-01-01.
```

## sort_values() method

```python
(data['team1'].value_counts() + data['team2'].value_counts()).sort_values() # default ascending=True, set it to false for descending.
```

```python
(data['team1'].value_counts() + data['team2'].value_counts()).sort_values(ascending=False).head()
```

```python
# sorting this data on the basis of a column.
# here the dataframe is sorted on the basis of city column.
data.head(3)
```

```python
data.sort_values('city')
# although these operations don't make permanent changes to the data, we can make them permanent by passing the inplace=True parameter.
```

```python
# We can also sort the data based on multiple columns by passing a list of columns as an argument.

data.sort_values(['city','date'])
```

```python
data.sort_values(['city','date'],ascending=[True,False]) #Pulls all the city names in ascending order and latest played match on the top.
```

## drop_duplicates() function

```python
data.drop_duplicates(subset=['city']) # Pulls all the unique city names.
```

```python
data.drop_duplicates(subset=['city','season']) # Like an and operation. Here cities with similar seasons are removed.
```

```python
data.drop_duplicates(subset=['city','season']).shape
```

```python
data.drop_duplicates(subset=['season'],keep='last')[['season','winner']].sort_values('season')
```
